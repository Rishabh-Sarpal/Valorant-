```python
print("===============================================")
print("      VALORANT AGENT EXPERT SYSTEM (AI)")
print("===============================================")

print("\nAnswer the following questions with yes or no.\n")

aggressive = input("Do you enjoy aggressive entry fights? ").lower()
support = input("Do you enjoy healing or supporting teammates? ").lower()
intel = input("Do you enjoy gathering enemy information? ").lower()
defense = input("Do you like defending sites using traps? ").lower()
smokes = input("Do you enjoy controlling areas with smokes/walls? ").lower()
mobility = input("Do you enjoy fast movement abilities? ").lower()

# Knowledge Base
agents = {

"Jett":{"role":"Duelist","abilities":["Cloudburst","Updraft","Tailwind","Blade Storm"],"score":0},
"Reyna":{"role":"Duelist","abilities":["Leer","Devour","Dismiss","Empress"],"score":0},
"Phoenix":{"role":"Duelist","abilities":["Blaze","Curveball","Hot Hands","Run It Back"],"score":0},
"Raze":{"role":"Duelist","abilities":["Boom Bot","Blast Pack","Paint Shells","Showstopper"],"score":0},
"Yoru":{"role":"Duelist","abilities":["Fakeout","Blindside","Gatecrash","Dimensional Drift"],"score":0},
"Neon":{"role":"Duelist","abilities":["Relay Bolt","Fast Lane","High Gear","Overdrive"],"score":0},
"Iso":{"role":"Duelist","abilities":["Contingency","Double Tap","Undercut","Kill Contract"],"score":0},

"Sage":{"role":"Sentinel","abilities":["Healing Orb","Barrier Orb","Slow Orb","Resurrection"],"score":0},
"Cypher":{"role":"Sentinel","abilities":["Trapwire","Cyber Cage","Spycam","Neural Theft"],"score":0},
"Killjoy":{"role":"Sentinel","abilities":["Nanoswarm","Alarmbot","Turret","Lockdown"],"score":0},
"Chamber":{"role":"Sentinel","abilities":["Trademark","Headhunter","Rendezvous","Tour De Force"],"score":0},
"Deadlock":{"role":"Sentinel","abilities":["GravNet","Sonic Sensor","Barrier Mesh","Annihilation"],"score":0},

"Brimstone":{"role":"Controller","abilities":["Stim Beacon","Incendiary","Sky Smoke","Orbital Strike"],"score":0},
"Omen":{"role":"Controller","abilities":["Shrouded Step","Paranoia","Dark Cover","From The Shadows"],"score":0},
"Viper":{"role":"Controller","abilities":["Snake Bite","Poison Cloud","Toxic Screen","Viper's Pit"],"score":0},
"Astra":{"role":"Controller","abilities":["Gravity Well","Nova Pulse","Nebula","Cosmic Divide"],"score":0},
"Harbor":{"role":"Controller","abilities":["Cascade","Cove","High Tide","Reckoning"],"score":0},
"Clove":{"role":"Controller","abilities":["Meddle","Pick-me-up","Ruse","Not Dead Yet"],"score":0},

"Sova":{"role":"Initiator","abilities":["Shock Bolt","Owl Drone","Recon Bolt","Hunter's Fury"],"score":0},
"Breach":{"role":"Initiator","abilities":["Aftershock","Flashpoint","Fault Line","Rolling Thunder"],"score":0},
"Skye":{"role":"Initiator","abilities":["Regrowth","Trailblazer","Guiding Light","Seekers"],"score":0},
"KAYO":{"role":"Initiator","abilities":["FLASH/drive","FRAG/ment","ZERO/point","NULL/cmd"],"score":0},
"Fade":{"role":"Initiator","abilities":["Seize","Haunt","Prowler","Nightfall"],"score":0},
"Gekko":{"role":"Initiator","abilities":["Wingman","Dizzy","Mosh Pit","Thrash"],"score":0}
}

# Inference Engine
for agent in agents:

    role = agents[agent]["role"]

    if aggressive == "yes" and role == "Duelist":
        agents[agent]["score"] += 3

    if support == "yes" and agent in ["Sage","Skye","Clove"]:
        agents[agent]["score"] += 3

    if intel == "yes" and role == "Initiator":
        agents[agent]["score"] += 3

    if defense == "yes" and role == "Sentinel":
        agents[agent]["score"] += 3

    if smokes == "yes" and role == "Controller":
        agents[agent]["score"] += 3

    if mobility == "yes" and agent in ["Jett","Neon","Yoru"]:
        agents[agent]["score"] += 2


# Find Best Agents
best_score = max(agent["score"] for agent in agents.values())

best_agents = [agent for agent in agents if agents[agent]["score"] == best_score]

print("\nAnalyzing playstyle...\n")

print("Recommended Agents:")
for agent in best_agents:
    print("-", agent)

print("\nAgent Details\n")

for agent in best_agents:
    print("Agent:", agent)
    print("Role:", agents[agent]["role"])
    print("Abilities:")
    for ability in agents[agent]["abilities"]:
        print("  •", ability)
    print()
```
